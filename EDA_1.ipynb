### NPS

0 - 6: Detratores <br>
7 - 8: Neutros <br>
9 - 10: Promotores <br>

NPS = % promotores - % detratores

```python
px.histogram(data_import['nps_grade'])
```

```python
nps_count = data_import['nps_grade'].astype(int)
nps_count.unique()
```

```python
pd.interval_range(start=0, end=10)
IntervalIndex=([0,6], [7,8], [9,10])

pd.cut(nps_count, bins=IntervalIndex)
```
